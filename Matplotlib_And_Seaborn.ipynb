```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
import seaborn as sns
```

```
iris = sns.load_dataset('iris')
iris.head()
```

```
plt.plot(iris['sepal_length'])
plt.show()
```

```
plt.plot(iris['sepal_length'],color='red')
plt.title('Iris Sepal Length')
plt.xlabel('Indexes')
plt.ylabel('Sepal Length')
plt.show()
```

```
plt.figure(figsize=(6,4))
plt.plot(iris['sepal_length'],color='red',linestyle='--')
plt.title('Iris Sepal Length')
plt.xlabel('Indexes')
plt.ylabel('Sepal Length')
plt.show()
```

```
plt.figure(figsize=(6,4))
plt.plot(iris['sepal_length'],color='red',marker='x')
plt.title('Iris Sepal Length')
plt.xlabel('Indexes')
plt.ylabel('Sepal Length')
plt.show()
```

```
plt.figure(figsize=(6,4))
plt.plot(iris['sepal_length'],color='red',label='sepal length')
plt.plot(iris['sepal_width'],color='green',label = 'sepal width')
plt.title('Iris Sepal Length')
plt.xlabel('Indexes')
plt.ylabel('Length Width')
plt.legend(loc='best')
plt.show()
plt.savefig('iris')
```

```
tip = sns.load_dataset('tips')
tip.head()
```

```
plt.bar(tip['day'],tip['total_bill'],color='red')
plt.title('Total bill on Different days',fontsize=20)
plt.show()
```

```
plt.barh(tip['day'],tip['total_bill'],color='red')
plt.title('Total bill on Different days',fontsize=20)
plt.show()
```

```
plt.style.use('fivethirtyeight')
plt.hist(tip['tip'],bins=10,edgecolor='black')
plt.show()
```

```
plt.boxplot(tip['total_bill'])
plt.show()
```

```
data_x = tip.groupby('day')['total_bill'].sum().reset_index()
data_x
```

```
x = data_x['day']
y = data_x['total_bill']
```

```
explode = [0,0,0,0.1]
plt.pie(y,labels=x,autopct='%1.1f%%',colors=['red','green','yellow','blue'],explode=explode)
plt.show()
```
